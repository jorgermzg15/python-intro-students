# 4. Estructuras de control

````python
"""
Condicionales 'if statements'
Los principales signos comparativos son:
Menor que <
Mayor que >
Igual que ==
Menor igual <=
Mayor igual >=
"""
````

````python
# Otras formas de 'validar' información de un objeto con estructura de datos
# Ejemplo de listas.
if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")

if "manzana" in frutas:
    print("si")
else:
    print("no")
````

````python
# Bucles
# Ejemplo de un rango a con una lista hecha a mano.

for i in [0,1,2,3,4,5]:
    print(i)
````

````python
"""
Ejemplo de un bucle utilizando función range()
  Nota: En Python, los índices comienzan en 0 y la función range() excluye el valor final.
  Esto es importante tener en cuenta al trabajar con bucles y acceso a elementos en listas o arrays.
  Python sigue la convención de la notación de intervalos en matemáticas y ciencias de la computación [inicio, fin)
"""

for i in range(6):
    print(i)
    
````

````python
# Ejemplo con lista
for f in frutas:
    print(f)
````

````python
# Ejemplo combinado
# combinaremos una condicional con un bucle y utilizaremos un "format string" y variables.

fruta_favorita = "manzana"

for fruta in frutas:
    if fruta == fruta_favorita:
        print(f'La {fruta} es mi fruta favorita')
    else:
        print(f'La {fruta} no es mi favorita')
````

# 5. Introducción a Librerías.

### Fundamentos de Programación
1.  **Variable**: Un nombre que se asigna a un valor en un programa. Las
variables pueden almacenar diferentes tipos de datos, como números, texto o estructuras de datos más complejas.
  - `nombre = 'Jorge'`

2.  **Función**: Un bloque de código que realiza una tarea específica y puede ser reutilizado en diferentes partes del programa. Las funciones pueden recibir argumentos y devolver valores.
  - `print()`

3.  **Clase**: Una plantilla para crear objetos que define propiedades y comportamientos. Las clases son fundamentales en la programación orientada a objetos y permiten la creación de objetos personalizados.
  - `str()`, contiene métodos como `.upper()` y `.lower()`
  ```
  nombre = str('Jorge')
  print(nombre.lower())
  ```
  `[out]: 'jorge'`

4.  **Método**: Una función que se define dentro de una clase y se utiliza para realizar operaciones en objetos de esa clase. Los métodos pueden acceder y modificar los atributos de un objeto.
  - `.append()` de la clase `list` que agrega un elemento al final de una lista
    ```
    lista = ['Jorge', 'Ramírez']
    lista.append('García')
    print(lista)
    ```
    `[out]: ['Jorge', 'Ramírez', 'García']`

5.  **Librería** (o biblioteca): Un conjunto de código preescrito que proporciona funcionalidades específicas para ser utilizadas en programas. Las librerías suelen incluir funciones, clases y variables que pueden ser importadas y utilizadas en diferentes proyectos.
  - la librería `math` que proporciona funciones para cálculos matemáticos como `sin()` y `cos()`
  ```
  import math
  result = math.sin(10)
  print(result)
  ```
  `[out]: -0.5440211108893698`

6.  **Paquete**: Un conjunto de librerías o módulos relacionados que se distribuyen juntos. Los paquetes pueden incluir varias librerías, documentación y ejemplos de uso.
  - El paquete `numpy` proporciona herramientas para estructurar y analizar datos.
  ```
  import numpy
  x = np.array([1, 2, 3]) # Creación de un vector.
  suma = np.sum(x)
  print(suma)
  ```
  `[out]: 6`
